Execute esta c√©lula para verificar se o ambiente est√° configurado corretamente antes de executar os notebooks do pipeline.

```python
import os
import sys
from pathlib import Path

# Verificar diret√≥rio atual
current_dir = Path().resolve()
print(f"üìÅ Diret√≥rio atual de trabalho: {current_dir}")
print()

# Tentar encontrar raiz do projeto
if 'notebooks' in str(current_dir):
    project_root = current_dir.parent
    print(f"‚úÖ Projeto raiz detectado: {project_root}")
elif (current_dir / 'scripts').exists():
    project_root = current_dir
    print(f"‚úÖ Projeto raiz detectado (estamos na raiz): {project_root}")
else:
    # Tentar caminho absoluto padr√£o
    project_root = Path('/Users/larissa/Desktop/TCC_metatrascriptomica')
    if project_root.exists():
        print(f"‚úÖ Projeto raiz encontrado (caminho absoluto): {project_root}")
    else:
        print(f"‚ùå Projeto n√£o encontrado em {project_root}")
        project_root = None

if project_root:
    print()
    print("üîç Verificando estrutura do projeto:")
    
    # Verificar diret√≥rios essenciais
    checks = {
        'scripts/': project_root / 'scripts',
        'data/': project_root / 'data',
        'notebooks/': project_root / 'notebooks',
        'results/': project_root / 'results'
    }
    
    for name, path in checks.items():
        status = '‚úÖ' if path.exists() else '‚ùå'
        print(f"  {status} {name}: {path}")
    
    # Verificar scripts Python
    print()
    print("üìù Verificando scripts Python:")
    scripts_dir = project_root / 'scripts'
    if scripts_dir.exists():
        scripts = list(scripts_dir.glob('*.py'))
        print(f"  ‚úÖ Encontrados {len(scripts)} scripts:")
        for script in sorted(scripts):
            print(f"     - {script.name}")
    else:
        print(f"  ‚ùå Diret√≥rio scripts n√£o encontrado")
    
    # Adicionar ao path
    print()
    if str(project_root) not in sys.path:
        sys.path.insert(0, str(project_root))
        print(f"‚úÖ Projeto adicionado ao sys.path: {project_root}")
    else:
        print(f"‚ÑπÔ∏è Projeto j√° est√° no sys.path")
    
    # Testar importa√ß√£o
    print()
    print("üß™ Testando importa√ß√µes:")
    try:
        import scripts
        print("  ‚úÖ import scripts - OK")
    except ImportError as e:
        print(f"  ‚ùå import scripts - FALHOU: {e}")
    
    try:
        from scripts.taxonomic_classification import run_kraken2
        print("  ‚úÖ from scripts.taxonomic_classification import run_kraken2 - OK")
    except ImportError as e:
        print(f"  ‚ùå Importa√ß√£o de fun√ß√£o - FALHOU: {e}")
    
    print()
    print("‚úÖ Configura√ß√£o do ambiente verificada!")
    print(f"üí° Para usar no notebook, execute: os.chdir('{project_root / 'notebooks'}') ou use project_root definido acima")
```
